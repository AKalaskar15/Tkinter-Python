## Databases

```python
from tkinter import *
from PIL import ImageTk,Image
import sqlite3
```

```python
root=Tk()
root.title("Creating Database")
root.geometry("400x400")

# Creating database or connecting to database
conn=sqlite3.connect('address_book.db')

# Create cursor
c=conn.cursor()

# Create table
c.execute(""" CREATE TABLE ADDRESSES (
              first_name text,
              last_name text,
              street text,
              city text,
              state text,
              telephone_no integer,
              pincode integer
)""")


# Commit changes to database
conn.commit()

# Close connection
conn.close()

root.mainloop()
```
